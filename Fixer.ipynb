```python
df = pd.read_csv('/users/baribowo/Documents/maintopic/HxOy_view_3.csv', delimiter="\t")
df["type"]
```

```python
def getOy(string):
    if type(string) == str:
        last_idx = len(string) - 1
        for i, st in enumerate(string):
            if st == "O":
                if i == last_idx:
                    return "1"
                elif i < last_idx:
                    if not (string[i+1].isalpha() or string[i+1] in [",",")","+","-"," "]) or string[i+1] in ["n","x"]:
                        return str(string[i+1])
                    elif string[i+1] == ")":
                        if i+1 == last_idx:
                            return "1"
                        else:
                            return str(string[i+2])
                    else:
                        return "1"

                    
print(getOy("(H2O)3O"))
                    
multips = []
for typ in df["type"]:
    multips.append(getOy(typ))
print(multips)
df["y"] = multips
df = df.sort_values('y')
print(df)
```

```python
df.to_csv('/users/baribowo/Documents/maintopic/HxOy_view_3_sorted_by_y.csv', index=False, sep="\t")
```
